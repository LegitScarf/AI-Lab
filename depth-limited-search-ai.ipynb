```python
def dfs_limited(graph, node, visited, depth, limit):
    if node not in visited and depth<=limit:
        print(node, end=" ")
        visited.add(node)
        for neighbor in graph[node]:
            dfs_limited(graph, neighbor, visited, depth+1, limit)

graph = {
    'A':['B', 'C'],
    'B':['D', 'E'],
    'C':['F'],
    'D':[],
    'E':['F'],
    'F':[]
}

print(graph)

print("DFS Traversal with Depth Limit:")
visited_set = set()
depth_limit = 2
dfs_limited(graph, 'A', visited_set, 0, depth_limit)
```
